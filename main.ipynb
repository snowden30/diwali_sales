```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.shape
```

```python
#only head will give by default 5 rows else you can give as parameter  
df.head(10)
```

```python
df.info()
```

```python
#removes the column name status and unnamed1  axis=1 means delete the whole column,
#inplace false ->This creates a new DataFrame without modifying the original df
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
# checking no of null values
pd.isnull(df).sum()
```

```python
#it will delete entire rows where any of the values are null.
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
#amount is in float now we will change its data type to int 
df['Amount']=df['Amount'].astype('int')
```

```python
df.info()
```

```python
#to get basic sum min max of all columns
df.describe()
```

```python
# for specific columns 
df[['Age','Amount']].describe()
```

```python
# sns.countplot(x='Gender',data=df) will give the plot but wont give the values of the plot 
# for loop
ax=sns.countplot(x='Gender',data=df)
for mybar in ax.containers:
    ax.bar_label(mybar)
```

```python
#if hue was not there now only acc to age group it will be made
ax=sns.countplot(x='Age Group',hue='Gender',data=df)
for mybar in ax.containers:
    ax.bar_label(mybar)
```

```python
#states with most orders  as_index=False by default its true which makes state as index,  head 10 is to get top 10 states 
ax=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
#to increase the size of graph as the names were overlapping first is width then is the height
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x='State',y='Orders',data=ax)
```

```python
#orders were more from kerala but the amount was less
ax=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x='State',y='Amount',data=ax)
```

```python
# according to marital status most shopping is from married men
# married means 0 and unmarried is 1
ax=sns.countplot(data=df,x='Marital_Status',hue='Gender')
sns.set(rc={'figure.figsize':(10,5)})
for mybar in ax.containers:
    ax.bar_label(mybar)
```

```python
#to see which sector people are shopping most
ax=sns.countplot(data=df,x='Occupation')
sns.set(rc={'figure.figsize':(20,5)})
for mybar in ax.containers:
    ax.bar_label(mybar)
```

```python
#most money spent sector wise
ax=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=ax,x='Occupation',y='Amount')
```

```python
# what kind of products were bought most
ax=sns.countplot(data=df,x='Product_Category')
for mybar in ax.containers:
    ax.bar_label(mybar)
# so mostly people buy clothes then food then gadgets
```

```python
# what products they spent most,
ax=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(data=ax,x='Product_Category',y='Amount')
# although most peopel bought clothes they spent most on food
```

```python
#so final conclusion is that married women age 26-35 from up, maharashtra and karnataka working in
 # IT,healthcare and aviation are more likely to buy food, clothing and electronics
```
